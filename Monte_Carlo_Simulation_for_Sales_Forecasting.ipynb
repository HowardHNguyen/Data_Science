## Assume:

1. Baseline sales: $1,000,000
2. Factors influencing sales:
 - Marketing Spend Effectiveness: Increase in sales due to marketing efforts.
 - Economic Growth Impact: Percentage change in sales due to economic conditions.
 - Customer Adoption Rate: Increase in sales due to customer acquisition.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Set baseline sales and probability distributions for each variable.
# Baseline sales
baseline_sales = 1_000_000

# Define distributions for uncertain parameters
marketing_effectiveness = np.random.normal(loc=0.1, scale=0.02, size=10_000)  # 10% average increase with 2% standard deviation
economic_growth = np.random.normal(loc=0.03, scale=0.01, size=10_000)         # 3% average growth with 1% standard deviation
customer_adoption = np.random.normal(loc=0.05, scale=0.015, size=10_000)      # 5% average increase with 1.5% standard deviation
```

Simulate Sales Scenarios - Calculate sales for each iteration by combining the random draws from the distributions.

```python
# Monte Carlo simulation: 10,000 iterations
simulated_sales = baseline_sales * (1 + marketing_effectiveness + economic_growth + customer_adoption)

# Summary statistics
mean_sales = np.mean(simulated_sales)
std_sales = np.std(simulated_sales)
percentile_5 = np.percentile(simulated_sales, 5)
percentile_95 = np.percentile(simulated_sales, 95)

print(f"Mean Sales: ${mean_sales:,.2f}")
print(f"Standard Deviation: ${std_sales:,.2f}")
print(f"5th Percentile: ${percentile_5:,.2f}")
print(f"95th Percentile: ${percentile_95:,.2f}")
```

Visualize the Results - Plot the distribution of simulated sales.

```python
# Plotting the sales distribution
plt.figure(figsize=(10, 6))
plt.hist(simulated_sales, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean_sales, color='red', linestyle='--', label=f'Mean Sales: ${mean_sales:,.2f}')
plt.axvline(percentile_5, color='green', linestyle='--', label=f'5th Percentile: ${percentile_5:,.2f}')
plt.axvline(percentile_95, color='orange', linestyle='--', label=f'95th Percentile: ${percentile_95:,.2f}')
plt.title('Monte Carlo Simulation: Sales Forecast Distribution')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### These results from the Monte Carlo simulation indicate the following:

1. Mean Sales: The expected average sales value is approximately $1,179,771.10. This represents the central tendency of all simulated outcomes and provides a solid reference for forecasting purposes.

2. Standard Deviation: The variability or spread of sales forecasts is $26,997.94, indicating that most sales outcomes will deviate by about $27,000 from the mean value. This relatively low standard deviation suggests consistent and reliable forecasting results.

3. 5th Percentile: At the lower end, 5% of sales outcomes are less than $1,134,885.71. This value provides a conservative estimate of the minimum expected sales, which is useful for planning under worst-case scenarios.

4. 95th Percentile: At the upper end, 5% of sales outcomes exceed $1,224,063.72. This represents the optimistic boundary, offering insight into best-case scenarios.

### Interpretation of Results
The simulation confirms that sales are expected to remain stable within the range of $1,134,885.71 to $1,224,063.72 for 90% of the scenarios. This narrow range of outcomes reflects the high confidence in the forecast and underscores the strength of incorporating probabilistic methods for planning.

The distribution curve in the histogram shows a bell shape, which is consistent with the normal distribution assumptions used in this Monte Carlo simulation. The plotted lines for the mean, 5th, and 95th percentiles further validate the scenario-based decision-making framework.

### Recommendations
1. Scenario Planning: Use the 5th and 95th percentiles to plan for conservative and optimistic sales scenarios. Adjust inventory, staffing, and marketing budgets accordingly to accommodate these ranges.
2. Risk Mitigation: The narrow standard deviation allows for targeted strategies to manage deviations from the mean.
3. Optimization: Explore adjusting input variables (e.g., marketing spend, customer acquisition rate) to shift the mean sales value upward or reduce variability.

This simulation provides actionable insights, offering robust support for both strategic planning and risk assessment. If additional factors such as seasonality or external market conditions are known, they can be integrated into the model for even more refined forecasting.
